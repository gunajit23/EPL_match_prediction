```python
import numpy as np
import pandas as pd
```

# Importing data

```python
data_home = pd.read_csv("data_home.csv")
data_away = pd.read_csv("data_away.csv")
```

```python
data_home.head()
```

```python
data_away.head()
```

# Removing unwanted parts of data

```python
col = ["xG","xGA","xPTS"]
for value in col:
    data_home[value] = data_home[value].str.split("+").str[0]
    data_home[value] = data_home[value].str.split("-").str[0]
    data_away[value] = data_away[value].str.split("+").str[0]
    data_away[value] = data_away[value].str.split("-").str[0]
    
```

# Calculating stats per game

```python
data_home["NPxG_pg"] = round((data_home["NPxG"]/data_home["M"]),2)
data_home["NPxGA_pg"] = round((data_home["NPxGA"]/data_home["M"]),2)
data_away["NPxG_pg"] = round((data_away["NPxG"]/data_away["M"]),2)
data_away["NPxGA_pg"] = round((data_away["NPxGA"]/data_away["M"]),2)
```

```python
data_home.set_index("Team",inplace=True)
data_away.set_index("Team",inplace=True)
```

# Importing fixtures

```python
fix = pd.read_csv("GW9.csv")
```

```python
pred = fix.copy()
```

```python
pred
```

# Relevant stats for each game

```python
for i in range(0,10):
    pred.loc[i,"xG_home"] = (data_home.loc[pred.loc[i]["Home"]]["NPxG_pg"])
    pred.loc[i,"xGA_home"] = (data_home.loc[pred.loc[i]["Home"]]["NPxGA_pg"])
    pred.loc[i,"xG_away"] = (data_away.loc[pred.loc[i]["Away"]]["NPxG_pg"])
    pred.loc[i,"xGA_away"] = (data_away.loc[pred.loc[i]["Away"]]["NPxGA_pg"])
    
```

# Getting predictions

```python
for i in range(0,10):
    pred.loc[i,"xGS"] = round((((pred.loc[i,"xG_home"]*pred.loc[i,"xGA_away"]/pred["xGA_away"].mean())+(pred.loc[i,"xG_home"]*pred.loc[i,"xGA_away"]/pred["xG_home"].mean()))/2),2)
    pred.loc[i,"xGC"] = round((((pred.loc[i,"xGA_home"]*pred.loc[i,"xG_away"]/pred["xG_away"].mean())+(pred.loc[i,"xGA_home"]*pred.loc[i,"xG_away"]/pred["xGA_home"].mean()))/2),2)
    
```

```python
pred['Score'] = pred.xGS.map(str) + "-" + pred.xGC.map(str)
pred.drop(["xG_home","xGA_home","xG_away","xGA_away","xGS","xGC"],axis=1,inplace=True)
pred = pred[["Home","Score","Away"]]
```
